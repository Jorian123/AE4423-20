Gravity Model:
    $D_{ij} = k\frac{(pop_{i}pop_{j})^{b1}(GDP_{i}GDP_{j})^{b2}}{(f\cdot d_{ij})^{b3}}$

Linear formula using logarithms: $\ln (D_{ij}) = \ln (k)+b1\cdot \ln (pop_ipop_j)+b2\cdot \ln(GDP_iGDP_j)-b3\cdot \ln(f\cdot d_{ij})$

```python
col= []
```

```python
for i in demand.index:
    for j in demand.index:
        if i!=j:
            col.append(i + "-" +j)
```

```python
d_set = pd.DataFrame(index=col)
```

```python
for i in demand.index:
    for j in demand.index:
        if i!=j:
            d_set.loc[i + "-" + j,"ln_Demand"]=math.log(demand.loc[i, j])
for i in demand.index:
    for j in demand.index:
        if i!=j:
            d_set.loc[i + "-" + j,"ln_pop"]=math.log(air_info[air_info["ICAO Code"]==i]["Population"].values*air_info[air_info["ICAO Code"]==j]["Population"].values)
for i in demand.index:
    for j in demand.index:
        if i!=j:
            d_set.loc[i + "-" + j,"ln_GDP"]=math.log(air_info[air_info["ICAO Code"]==i]["GDP"].values*air_info[air_info["ICAO Code"]==j]["GDP"].values)
f = 1.42/1.5
for i in demand.index:
    for j in demand.index:
        if i!=j:
            d_set.loc[i + "-" + j,"ln_fuel"]=-math.log(f*distan.loc[i, j])
```
